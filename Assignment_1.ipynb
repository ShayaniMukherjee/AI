```python
#imported libraries like pandas(for data analysis), numpy(to do mathematical funtions with array), matplotlib(to visualize the data by ploting them)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#reading the data file
data = pd.read_csv('dataset_agriculture_01.csv') 
```

```python
#distribution of the data in the data shape
data.shape
```

```python
#description of the data
data.describe()
```

```python
#returning the total of the number of misssing values in the datasets
data.isnull().sum()
```

```python
#data cleaning
data['Temparature'].fillna(data['Temparature'].mode()[0], inplace=True)
data['Humidity'].fillna(data['Humidity'].mode()[0], inplace=True)
data['Wind_Speed'].fillna(data['Wind_Speed'].mode()[0], inplace=True)
data['Surface_Pressure'].fillna(data['Surface_Pressure'].mode()[0], inplace=True)
data['Profile_Soil_Moisture'].fillna(data['Profile_Soil_Moisture'].mode()[0], inplace=True)
```

```python
#checking the null values
data.isnull().sum()
```

```python
#copy
data_org=data
```

```python
#Percentile on original, not updated column value

Humi_1 = np.percentile(data['Humidity'], 20)
Humi_2 = np.percentile(data['Humidity'], 80)
Humi_1, Humi_2
```

```python
data_Backup = data
data = data[(data['Humidity'] >= Humi_1) & (data['Humidity'] <= Humi_2)] #Removing outliers
```

```python
#printing the ditribution of the data
data.shape
```

```python
data_org
```

```python
#imported ExtraTreesClassifier from sklearn.ensemble library
from sklearn.ensemble import ExtraTreesClassifier
```

```python
#Feature Importance Score
X = data_org.iloc[:,0:8] #independent columns
y = data_org.iloc[:,-1] # pick last column for the target feature
model = ExtraTreesClassifier()
model.fit(X,y)
print(model.feature_importances_)
#feature_importances of tree based classifiers
#plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(3).plot(kind='barh')
plt.show()
```

```python
#imported SelectKBest and chi2 from sklearn library
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
#applying SelectKBest class to extract top 3 best features

bestfeatures = SelectKBest(score_func=chi2, k=3)
fit = bestfeatures.fit(X,y)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(X.columns)
scores = pd.concat([dfcolumns,dfscores],axis=1)
scores.columns = ['specs','score']
print(scores.nlargest(3,'score')) #printing the 3 best features
```

```python
#reading the data file
data = pd.read_csv('drug_class.csv')
```

```python
data
```

```python
#distribution of the data in the data shape
data.shape
```

```python
#checking the total of the number of misssing values in the datasets
data.isnull().sum()
```

```python
print(data.Sex.unique(),data.BP.unique(),data.Cholesterol.unique(),data.Drug.unique())
```

```python
data["Sex"] = data.Sex.replace('F',1)
data["Sex"] = data.Sex.replace('M',0)

data["BP"] = data.BP.replace('HIGH',2)
data["BP"] = data.BP.replace('NORMAL',1)
data["BP"] = data.BP.replace('LOW',0)

data["Cholesterol"] = data.Cholesterol.replace('HIGH',1)
data["Cholesterol"] = data.Cholesterol.replace('NORMAL',0)

data["Drug"] = data.Drug.replace('drugA',4)
data["Drug"] = data.Drug.replace('drugB',3)
data["Drug"] = data.Drug.replace('drugC',2)
data["Drug"] = data.Drug.replace('drugX',1)
data["Drug"] = data.Drug.replace('DrugY',0)
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
# Split data into training and test data
X = data.iloc[:,0:5].values
y = data.iloc[:,5].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Create the logistic regression classifier
logmodel = LogisticRegression(max_iter=350)
# Train the logmodel
logmodel.fit(X_train,y_train)
predictions = logmodel.predict(X_test)
# Computing accuracy
accuracy=confusion_matrix(y_test,predictions)
accuracy=accuracy_score(y_test,predictions)
print("Accuracy of the Logistic Regression model is: ",round(accuracy*100,2),"%")
```

```python
#Import Gaussian Naive Bayes model
from sklearn.naive_bayes import GaussianNB
```

```python
#create Naïve Bayes classifier
gnb = GaussianNB()
#train the classifier
gnb.fit(X_train, y_train)
# Predict the values for training data
y_pred = gnb.predict(X_test)
# Computing accuracy
accuracy=accuracy_score(y_test,predictions)
print("Accuracy of the Naïve Bayes classifier model is:",round(accuracy*100,2),"%")
```

```python
#reading the data file
data = pd.read_csv('kc_house_data.csv')
```

```python
#distribution of the data in the data shape
data.shape
```

```python
data
```

```python
#drop the unneccessary column
data = data.drop(['id', 'date'], axis=1)
```

```python
data
```

```python
# Training and testing the data
X = data.values[:, 1:19]
Y = data.values[:,0]
X_train, X_test, Y_train, Y_test = train_test_split( X, Y, test_size = 0.1, random_state = 100)
```

```python
from sklearn.linear_model import LinearRegression
import sklearn.metrics as sm
```

```python
# Create linear regressor object
linear_model = LinearRegression()
# Train the model using the training sets
linear_model.fit(X_train,Y_train)
# Predict the output
y_test_pred = linear_model.predict(X_test)
```

```python
print("R2 score =", round(sm.r2_score(Y_test, y_test_pred)*100, 2),"%")
```

```python
#Searching for RMSE value
from sklearn.metrics import mean_squared_error
rmse = mean_squared_error(Y_test, y_test_pred, squared=False)
print("RMSE value: ",rmse)
```
